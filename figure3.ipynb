```python
import BIOMD0000000206 as wolf
import tellurium as te

import biomodeltoolbox as bmt
from numpy import *
rlist = linspace(0, 4, 1000)

rr = te.loadSBMLModel(wolf.sbmlString)
```

```python
rr.resetToOrigin()
X = []
for k9 in rlist:
    rr.resetToOrigin()
    rr.k9 = k9
    rr.simulate(0, 2, 1000, ['na'])
    if len(X):
        X = hstack([X, rr.simulate(0, 5, 1000, ['na'])])
    else:
        X = hstack([rr.simulate(0, 5, 1000, ['na'])])
X = hsplit( vstack(X), rlist.size)

from scipy import stats
import matplotlib.pyplot as plt

H = map( lambda Z : stats.histogram( Z, numbins=300 )[0],X)
H = map( lambda Z : 1-Z/Z.max(), H )
H = vstack(H)

plt.imshow( rot90(H), aspect = 'auto' , extent = [0, 10, 0, 1])
plt.bone()
plt.xlabel('r')
plt.ylabel(r'$X_{n \rightarrow \infty}$')
plt.show()
```
